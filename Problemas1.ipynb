## 1. Represente gráficamente las siguientes señales en tiempo discreto:

```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin, pi

def plot(n, y, title): 
    plt.stem(n, y)#, use_line_collection=True)
    plt.xlabel('n')
    plt.ylabel('x[n]')
    plt.title(title)
    plt.show()
```

```python
n = np.array(range(-15, 15))
delta_n = np.array([1 if x==0 else 0 for x in n])
delta_n1 = np.array([1 if x==1 else 0 for x in n])
y = delta_n - delta_n1
plot(n, y, "Problema 1a:  x[n] = δ[n] - δ[n-1]")
```

```python
u_nm1 = np.array([0 if x < 1 else 1 for x in n])
u_nM1 = np.array([0 if x < -1 else 1 for x in n])
y = u_nm1 - u_nM1
plot(n, y, "Problema 1b: x[n] = u[n-1] - u[n+1]")
```

```python
y = np.cos(2*pi*n)
plot(n, y, "Problema 1c: x[n] = cos(2*pi*n)")
```

```python
y = np.cos(2*pi*n/5)
plot(n, y, "Problema 1d: x[n] = cos(2*pi*n/5)")
```

```python
y = np.cos(n)
plot(n, y, "Problema 1e: x[n] = cos(n)")
```

```python
un = np.array([1 if x>= 0 else 0 for x in n])
y = np.exp(-n/4)*un
plot(n, y, "Problema 1f: x[n] = exp(-n/4)*u[n]")
```

```python
unm1 = np.array([1 if x>= 1 else 0 for x in n])
y = np.exp(-n/4)*unm1
plot(n, y, "Problema 1g: x[n] = exp(-n/4)*u[n-1]")
```

```python
unm10 = np.array([1 if x>= 10 else 0 for x in n])
y = np.cos(2*pi*n/3 + 1)*(un - unm10)
plot(n, y, "Problema 1h: x[n] = cos(2*pi*n/3+1) * (u[n]-u[n-10])")
```

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Create a discrete delta function (unit impulse)
n = 10  # Length of the signal
impulse = signal.unit_impulse(n, idx='mid')

# Plot the delta function
plt.stem(np.arange(n), impulse)#, use_line_collection=True)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Discrete Delta Function')
plt.show()
