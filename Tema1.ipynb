## 1.1 Señales discretas en el tiempo

$x: \mathbb{R} \rightarrow \mathbb{Z}$

o 

$x: \mathbb{R} \rightarrow \mathbb{C}$

```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin

x = range(-10, 10)
y = np.sin(x)/(1+np.abs(x))

plt.stem(x, y)#, use_line_collection=True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('y[n] = sin(n)/(|n|+1)')
plt.show()
```

###  Propiedades elementales

Transformaciones elementales:
|Propiedad|Descripción|
|:------|:-----------|
|Escalado por $\alpha \in  \mathbb{C}$|$ y[n] = \alpha x[n]$|
|Desplazamiento de $n_0 \in \mathbb{Z}$ muestras| $y[n] = x[n-n_0]$|
|Simetría especular| $y[n] = x[-n]$|
|Compresión por un factor $N \in \mathbb{Z}^+$|$y[n]=x[Nn]$|
